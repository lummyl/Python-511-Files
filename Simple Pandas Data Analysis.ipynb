```python
import pandas as pd
df = pd.read_excel("Bonneville Data.xlsx", "Bonneville 1-1-1960 to 12-31-20")
df
```

#### Question 1: What is the average number of sightings by species, total? 

```python
df.mean()
```

#### Question 2: What is the average number of sighting by species up until 1980?

```python
df['Date'] = pd.to_datetime(df.Date)
ts = pd.to_datetime('1/1/1981')
df.loc[df['Date'] < ts].mean()
```

#### Question 3: What were the max number sightings for each species in 1972?

```python
df.loc[df.Date.dt.year == 1972].describe()
```

#### Question 4: What days had the greatest sightings of Steelhead in 1960?

```python
df.loc[df.Date.dt.year == 1960].sort_values(['Steelhead'], ascending=False)
```
