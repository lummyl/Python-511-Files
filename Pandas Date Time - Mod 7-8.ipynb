### Working with time and dates in pandas

```python
import pandas as pd
df = pd.read_csv('7-8_uforeports.csv')
df
```

```python
df.dtypes
```

```python
df['Time'] = pd.to_datetime(df.Time)
```

```python
df
```

#### The code changes the object type of the time column. Some attributes you can pull out of the column now:

```python
df.Time.dt.hour.head()
```

```python
df.Time.dt.weekday_name.head()
```

#### Some operations you can perform:

```python
timestamp = pd.to_datetime('1/1/2000')
df.loc[df['Time'] >= timestamp]
```

```python
df.Time.max()- df.Time.min()
```

```python
%matplotlib inline
df['Year'] = df.Time.dt.year
df.head()
```
